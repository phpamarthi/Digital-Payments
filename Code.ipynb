```python
df = df[df.pref_method != 'Cash']
df = df[df.pref_method != 'Physical card']
df = df[df.pref_method != 'Cash;Physical card']
df.shape
```

```python
df['pref_method'] = df['pref_method'].str.replace('Cash;Physical card;Digital payment without digital card','digi_with_others')
df['pref_method'] = df['pref_method'].str.replace('Cash;Digital payment without digital card', 'digi_with_others')
df['pref_method'] = df['pref_method'].str.replace('Physical card;Digital payment without digital card','digi_with_others')
df['pref_method'] = df['pref_method'].str.replace('Digital payment without digital card','only_digi')
```

```python
df = df.sort_values(['Name'],ascending = [True])
```

```python
filter = df[df['Name'].str.contains('Bhaavya')]
filter = filter[['pref_method','risk','Aware','usage_for','advancements','ess_use','chances_fraud','mode_pay']]
filter
```

```python
df1 = df.assign(pref_method=df['pref_method'].str.split(';')).explode('pref_method')
df2 = df1.assign(risk=df1['risk'].str.split(';')).explode('risk')
df3 = df2.assign(Aware=df2['Aware'].str.split(';')).explode('Aware')
df4 = df3.assign(usage_for=df3['usage_for'].str.split(';')).explode('usage_for')
df5 = df4.assign(advancements=df4['advancements'].str.split(';')).explode('advancements')
df6 = df5.assign(ess_use=df5['ess_use'].str.split(';')).explode('ess_use')
df7 = df6.assign(chances_fraud=df6['chances_fraud'].str.split(';')).explode('chances_fraud')
df8 = df7.assign(mode_pay=df7['mode_pay'].str.split(';')).explode('mode_pay')
df8.reset_index(drop=True,inplace=True)
df8
```

```python
col_keep = df8.drop(['TimeStamp','Name', 'Age', 'suggestions'],axis = 1)
```

```python
dummy = pd.get_dummies(col_keep, drop_first = False)
dummy.reset_index(drop=True,inplace=True)
dummy
```

```python
dummy['pref_method_only_digi'] = dummy['pref_method_only_digi'].replace(1,10)
```

```python
dummy['pref_method'] = dummy['pref_method_only_digi']+dummy['pref_method_digi_with_others']
dummy['pref_method']
```

```python
#1 is pref_method_digi_with_others and 0 is pref_method_only_digi
dummy['pref_method'] = dummy['pref_method'].replace(10,0)
dummy['pref_method']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X= dummy.drop(['pref_method'], axis = 1)
y = dummy[['pref_method']]
```

```python
y.value_counts()
```

```python
##Segregating data for training and testing for Linear regression
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 0)
```

```python
y_test
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
##Creating dictionary to store results
d = {}
d["ModelAccuracy"] = model.score(X_test, y_test)
d["Intercept"] = model.intercept_
d
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, model.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
##Pulling predicted values
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)) 

print("Precision:",metrics.precision_score(y_test, y_pred)) 

print("Recall:",metrics.recall_score(y_test, y_pred)) 
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred) 
cnf_matrix 

class_names=['digi_with_others','only_digi'] # name  of classes 

fig, ax = plt.subplots() 

tick_marks = np.arange(len(class_names)) 

plt.xticks(tick_marks, class_names) 

plt.yticks(tick_marks, class_names) 

# create heatmap 

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 

ax.xaxis.set_label_position("top") 

plt.tight_layout() 

plt.title('Confusion matrix', y=1.1) 

plt.ylabel('Actual label') 

plt.xlabel('Predicted label')
```
